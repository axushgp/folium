```python
import folium
m=folium.Map(location=[12.950881335398455, 77.58618485084702], zoom_start=15)
m
```

```python
iconn=folium.Icon(color='black', icon="bezier-curve", prefix='fa')
folium.Marker(location=[12.950881335398455, 77.58618485084702], 
              tooltip="IDHAR CLICK KAR!!", 
              popup="<strong>LALBAGH GARDEN BROO!</strong>",
              icon=iconn).add_to(m)

m
```

```python
folium.Marker(
    location=[12.947409945148838, 77.5844253222362],
    tooltip="Click me!",
    popup="LalBagh Waterfall",
    icon=folium.Icon(color="lightgray", icon='crosshairs', prefix='fa'),
).add_to(m)

m
```
